1. **Гистограмма для числовых переменных:**

```python
import matplotlib.pyplot as plt
import seaborn as sns
ch_vars = ['Age', 'Annual Income ($)', 'Family Size', 'Work Experience']
num_cols = ch_vars
fig, axes = plt.subplots(1, len(num_cols), figsize=(6 * len(num_cols), 4))
if len(num_cols) == 1:
    axes = [axes]
for ax, col in zip(axes, num_cols):
    sns.histplot(df[col].dropna(), kde=True, ax=ax)
    ax.set_title(f'Распределение {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Частота')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

ch_vars = ['Age', 'Annual Income ($)', 'Family Size', 'Work Experience']

fig, axes = plt.subplots(1, len(ch_vars), figsize=(6*len(ch_vars), 4))
if len(ch_vars) == 1: axes = [axes]

for ax, col in zip(axes, ch_vars):
    sns.histplot(data=df, x=col, hue='Gender', kde=True,
                 stat='count', common_norm=False, alpha=0.45, ax=ax)
    ax.set_title(f'{col} (hue=Gender)')
    ax.set_xlabel(col); ax.set_ylabel('Частота')

plt.tight_layout(); plt.show()
```

2. **Столбчатая диаграмма для категориальной переменной `Class`:**

```python
sns.countplot(x='Gender', data=df)
plt.title('Распределение Gender')
plt.xlabel('Gender')
plt.ylabel('Количество')
plt.show()


top_n = 10
top_prof = df['Profession'].value_counts().head(top_n).index
df_top = df[df['Profession'].isin(top_prof)].copy()

sns.countplot(x='Profession', data=df_top,
              order=df_top['Profession'].value_counts().index)
plt.title(f'Распределение Profession (Top {top_n})')
plt.xlabel('Profession')
plt.ylabel('Количество')
plt.xticks(rotation=30)
plt.show()
```
