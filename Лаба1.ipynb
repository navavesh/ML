1. **Гистограмма для числовых переменных:**

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

# --- Списки переменных ---
num_v = ['Age', 'Annual Income ($)', 'Family Size']
cat_v = ['Gender', 'Profession', 'Spending Score (1-100)', 'Work Experience']

print("Гистограммы числовых переменных")
fig, axes = plt.subplots(1, len(num_v), figsize=(18, 4))

for ax, col in zip(axes, num_v):
    sns.histplot(df[col], kde=True, ax=ax, color='skyblue')
    ax.set_title(f'{col}')
    ax.set_xlabel("")
    ax.set_ylabel('Частота')

plt.tight_layout()
plt.show()

print("Распределения категориальных и порядковых переменных")
fig, axes = plt.subplots(2, 2, figsize=(14, 8))

for ax, col in zip(axes.flatten(), cat_v):
    sns.countplot(data=df, x=col, ax=ax, palette='pastel', hue=col, legend=False)
    ax.set_title(f'{col}', fontsize=11)
    ax.set_xlabel("")
    ax.set_ylabel("Количество", fontsize=9)
    ax.tick_params(axis='x', rotation=0, labelsize=8)

plt.tight_layout()
plt.show()

```
